```python
data["InvoiceDate"] = pd.to_datetime(data["InvoiceDate"])
```

```python
data["TotalSpending"] = data["Quantity"] * data["UnitPrice"]
```

```python
import datetime as dt

snapshot_date = max(data.InvoiceDate) + dt.timedelta(days=1)

#Calculating RFM metrics 
rfm = data.groupby(['CustomerID']).agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, 
    'InvoiceNo': 'count',
    'TotalSpending': 'sum' 
})
```

```python
# Renaming Columns
rfm.rename(columns={'InvoiceDate': 'Recency',
                    'InvoiceNo': 'Frequency',
                    'TotalSpending': 'MonetaryValue'}, inplace=True)
```

```python
rfm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dist. of Recency 
sns.distplot(rfm['Recency'])
plt.show()

# Dist. ofFrequency
sns.distplot(rfm['Frequency'])
plt.show()

# Dist. ofMonetaryValue
sns.distplot(rfm['MonetaryValue'])
plt.show()
```

```python
rfm
```

```python
# Renaming columns
rfm.columns = ['Recency', 'Frequency', 'MonetaryValue']

# Calculating RFM Scores

rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.qcut(rfm['MonetaryValue'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm
```

```python
# Combining RFM Scores and Creating the RFM Segment

rfm['RFM_Segment'] = rfm.apply(lambda x: str(x['R_Score']) + str(x['F_Score']) + str(x['M_Score']), axis=1) # yan yana yazar
rfm['RFM_Score'] = rfm[['R_Score', 'F_Score', 'M_Score']].sum(axis=1) #toplar
```

```python
rfm
```

```python
# Defining Customer Segments Based on RFM Scores

segt_map = {
    r'[1-2][1-2]': 'Hibernating',              # Hibernating
    r'[1-2][3-4]': 'At Risk',           # At Risk
    r'[1-2]5': 'Can’t Lose',               # Can’t Lose
    r'3[1-2]': 'About To Sleep',    # About To Sleep
    r'33': ' Need Attention',           # Need Attention
    r'[3-4][4-5]': 'Loyal Customers',       # Loyal Customers
    r'41': 'Promising',                  # Promising
    r'51': 'New Customers',                # New Customers
    r'[4-5][2-3]': 'Potential Loyalists',    # Potential Loyalists
    r'5[4-5]': 'Champions'                 # Champions
}
rfm['Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)
```

```python
rfm
```

```python
rfm['Segment'].value_counts()
```

```python
# Visualizing Segment Counts
sns.countplot(y='Segment', data=rfm, order = rfm['Segment'].value_counts().index)
plt.title('RFM Segment Dağılımı')
plt.xlabel('Sayısı')
plt.ylabel('Segment')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scaling The Data
scaler = StandardScaler()
rfm_numeric = rfm[['Recency', 'Frequency', 'MonetaryValue', 'R_Score', 'F_Score', 'M_Score', 'RFM_Score']]

rfm_scaled = scaler.fit_transform(rfm_numeric) 
```

```python
rfm_numeric
```

```python
rfm_scaled
```

```python
# K-means clusters
kmeans = KMeans(n_clusters=4, random_state=1)
```

```python
rfm_numeric['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
rfm_numeric
```

```python
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(rfm_scaled)
    sse[k] = kmeans.inertia_
```

```python
sse
```

```python
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

# V. Practical Applications <a class="anchor" id="practical-applications"></a>

## Impact on Business Decisions

RFM (Recency, Frequency, Monetary) analysis is a powerful tool for customer relationship management and marketing strategies. Businesses can use RFM analysis to better understand their customer base and make more effective decisions.

### Inventory Management

RFM analysis can be used to optimize inventory levels by considering the frequency of customer purchases and the most recent purchase time. For example:

- **High Frequency and Low Recency**: Stock levels of popular products should be kept high for these customers.
- **Low Frequency and High Recency**: Stock levels can be kept low for customers who purchase infrequently, reducing costs.

### Pricing Strategies

Dynamic pricing strategies can be developed based on the monetary value of customers:

- **High Monetary Value**: Special pricing and discounts can be offered to premium customers.
- **Low Monetary Value**: Promotions can be organized to encourage these customers to make more purchases.

### Customer Services

RFM segmentation allows for more efficient use of customer service resources:

- **High Frequency and Monetary Value**: VIP customer service can be prioritized for this group.
- **Low Frequency and Recency**: Automated services or self-service options may be more suitable for this group.

## Customer Lifecycle Management and Loyalty Programs

RFM analysis helps determine how to approach customers at various stages of the customer lifecycle:

- **New Customers (High Recency)**: Welcome offers and first purchase discounts can be provided.
- **Loyal Customers (High Frequency, High Monetary)**: These customers can be rewarded with loyalty programs and special offers.
- **Dormant Customers (Low Frequency)**: Reactivation campaigns can be used to win back these customers.

## Industry Success Stories

Examples of successful RFM analysis applications:

- **Retail**: A retail chain used RFM analysis to understand customer purchasing behavior and increased sales by 20% through targeted marketing campaigns.
- **E-Commerce**: An e-commerce platform used the RFM model for customer segmentation and increased customer loyalty by offering special offers to the most valuable customers.
- **Financial Services**: A bank improved customer interactions and increased cross-sell rates by 15% using RFM analysis.
